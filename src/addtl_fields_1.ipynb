### Article length: Fame Score

Get Wikipedia URL for getting fame score i.e. size of article:

Get page metadata:

```python
base_url_prefix = 'https://en.wikipedia.org/w/api.php?action=query&format=json&titles=' 
base_url_suffix = '&prop=revisions&rvprop=size'

def get_name_url(prefix, name_text, suffix):
    return prefix + name_text.replace(' ','_') + suffix

def get_page_size(wiki_name):
    try:
        size_url = get_name_url(base_url_prefix, wiki_name, base_url_suffix)
        size_page = requests.get(size_url)
        return size_page.json()['query']['pages'].values()[0]['revisions'][0]['size']
    except (KeyError, UnicodeDecodeError, ValueError):
        print wiki_name
```

### Birth and death data

Getting birthday data. It's probably helpful to see [Stack Overflow thread](https://stackoverflow.com/questions/12250580/parse-birth-and-death-dates-from-wikipedia)

```python
base_bday_prefix = 'https://en.wikipedia.org/w/api.php?action=query&format=json&titles='
base_bday_suffix = '&prop=revisions&rvprop=content&rvsection=0'

def is_number(elem):
    try:
        return type(int(elem)) == int
    except ValueError:
        return False

def get_only_ints(date_list):
    return [e for e in date_list if is_number(e) > 0]

def get_date_string(date_match):
    return "-".join(get_only_ints(date_match.groups()[0].strip(" ").split('|'))[:3])

life_death_counter = [0]
no_birth_death_urls = []

bday_re = re.compile('.*?\{{2}(?:B|b)irth (?:D|d)ate(.+?)\}{2}', re.MULTILINE|re.DOTALL)
dday_re = re.compile('.*?\{{2}(?:D|d)eath (?:D|d)ate(.+?)\}{2}', re.MULTILINE|re.DOTALL)

def get_life_death(name_text):
    
    # return value
    life_dates_out = ['', '', True]
    
    try:        
        # get page
        entry_url = get_name_url(base_bday_prefix, name_text, base_bday_suffix)
        entry_text = requests.get(entry_url).json()['query']['pages']
        entry_text = str(entry_text.values()[0]['revisions'][0]['*'].encode('utf-8'))

        # check for matches
        bday_match = re.match(bday_re, entry_text)
        dday_match = re.match(dday_re, entry_text)
        
        # extract dates if match
        if bday_match:
            life_dates_out[0] = get_date_string(bday_match)
        if dday_match:
            life_dates_out[1] = get_date_string(dday_match)
        
        # print url if no match found
        if bday_match or dday_match:
            life_dates_out[2] = False
    
    except (AttributeError, KeyError, ValueError) as e:
        print entry_url
        no_birth_death_urls.append(entry_url)
    
    return life_dates_out
```
